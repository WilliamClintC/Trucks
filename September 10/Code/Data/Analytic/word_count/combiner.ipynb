```python
import pandas as pd
import glob

# Define the pattern to match the CSV files
file_pattern = "Total_word_count_*.csv"

# Use glob to get a list of all matching files
csv_files = glob.glob(file_pattern)

# Initialize an empty list to store dataframes
dataframes = []

# Loop through each file and read it into a dataframe
for file in csv_files:
    df = pd.read_csv(file)
    dataframes.append(df)

# Concatenate all dataframes
combined_df = pd.concat(dataframes, ignore_index=True)
combined_df
```

```python
sorted_df = combined_df.sort_values(by='Location')
sorted_df
```

```python

# Write the combined dataframe to a new CSV file
sorted_df.to_csv("Combined_word_count.csv", index=False)
```
