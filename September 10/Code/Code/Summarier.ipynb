```python
import pandas as pd

df = pd.read_csv(r'C:\Users\clint\Desktop\Aug 26\Code Aug 28\Data\Raw\transposed_df.csv')
df
```

```python
df.set_index('Words', inplace=True)
df
```

```python
zero_columns = df.columns[df.eq(0).all(axis=0)]
zero_columns
```

```python
df.describe().loc[['count', 'mean', 'std', 'min', '50%']]
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is already defined and contains the data
plt.figure(figsize=(30, 30))  # Increase the figure size

# Use the tab20c colormap for better color differentiation
colors = plt.cm.tab20c.colors

ax = df.describe().loc[['count']].plot(kind='bar', color=colors)

# Customize the legend to be on the right side and have two columns
ax.legend(loc='center left', bbox_to_anchor=(1.5, 0.5), ncol=2)

# Add title and labels for better readability
plt.title('Count of Each Column')
plt.xlabel('Words')
plt.ylabel('Count')

# Adjust layout to fit the legend
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming df is already defined and contains the data
plt.figure(figsize=(40, 40))  # Increase the figure size

# Use the tab20c colormap for better color differentiation
colors = list(plt.cm.tab20.colors) + list(plt.cm.tab20.colors)[:24]

# Get the median values and sort them in descending order
median_values = df.describe().loc['50%'].sort_values(ascending=False)

# Plot the sorted median values
ax = median_values.plot(kind='bar', color=colors)

ax.set_ylim(0, 1100) 
# Customize the legend to be on the right side and have two columns
ax.legend(loc='center left', bbox_to_anchor=(1.5, 0.5), ncol=2)

# Add title and labels for better readability
plt.title('Median Word Count among all Locations ("Shall" bar cut for readability)', fontsize=50)
plt.xlabel('Median', fontsize=50)
plt.ylabel('Count', fontsize=50)

# Make the words on the x-axis and y-axis bigger
ax.set_xticklabels(ax.get_xticklabels(), fontsize=40)
ax.set_yticklabels(ax.get_yticks(), fontsize=40)

# Adjust layout to fit the legend
plt.tight_layout()

# Show the plot
plt.show()
```
