```python

# Function to merge similar columns
def merge_similar_columns(df):
    columns = df.columns
    merged_columns = {}
    
    for col in columns:
        base_col = col.rstrip('d')  # Remove trailing 'd' to find base column name
        if base_col in merged_columns:
            merged_columns[base_col].append(col)
        else:
            merged_columns[base_col] = [col]
    
    for base_col, similar_cols in merged_columns.items():
        if len(similar_cols) > 1:
            df[base_col] = df[similar_cols].sum(axis=1)
            df.drop(columns=[col for col in similar_cols if col != base_col], inplace=True)
    
    return df

# Apply the function to the DataFrame
cv_df = merge_similar_columns(cv_df)
print(cv_df)
```

```python

import numpy as np

# Assuming df is already loaded

def find_outliers_and_replace(df):
    for column in df.columns:
        mean = df[column].mean()
        std_dev = df[column].std()
        lower_bound = mean - 5 * std_dev
        upper_bound = mean + 5 * std_dev
        
        # Identify outliers
        outliers = (df[column] > upper_bound)
        
        # Replace non-outliers with NaN
        df[column] = df[column].where(outliers, np.nan)
    
    return df

# Apply the function to the DataFrame
df = find_outliers_and_replace(df)
df
```

```python
import pandas as pd
import folium
from folium import IFrame

# Assuming df_cleaned is already loaded
# Example: df_cleaned = pd.read_csv('path_to_your_file.csv')

# Initialize a map centered on the US
us_map = folium.Map(location=[37.0902, -95.7129], zoom_start=4)

for _, row in df_long_lat.iterrows():
    folium.CircleMarker(
        location=[row['INTPTLAT10'], row['INTPTLON10']],
        radius=5,
        color='red',
        fill=True,
        fill_color='red'
    ).add_to(us_map)

# Iterate through the first column to find non-NaN values
for index, value in df_cleaned.iloc[:, 0].dropna().items():
    # Get the latitude and longitude
    lat = df_cleaned.at[index, 'INTPTLAT10']
    lon = df_cleaned.at[index, 'INTPTLON10']
    
    # Get the column name
    column_name = df_cleaned.columns[0]
    
    # Create a popup with the index name, column name, and value
    popup_content = f"Index: {index}<br>Column: {column_name}<br>Value: {value}"
    popup = folium.Popup(IFrame(popup_content, width=200, height=100))
    
    # Add a marker to the map
    folium.Marker(location=[lat, lon], popup=popup).add_to(us_map)

# Save the map to an HTML file
us_map.save('us_map.html')
```
