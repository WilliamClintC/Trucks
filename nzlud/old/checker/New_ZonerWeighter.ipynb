```python
import pandas as pd

# Copy the first 4 rows of df
copy_df = df.head(4)

# Append the copied rows below
appended_df = pd.concat([copy_df] * 6, ignore_index=True)

appended_df.head()
```

```python
import re

def remove_numberings(text):
    # Use regex to remove the first occurrence of a number followed by a dot and a space, but not if followed by more digits
    return re.sub(r'\b\d+\.\s(?!\d)', '', text, count=1)

def replace_warehouse_with_office(text):
    return text.replace("warehouse", "office")

def replace_office_with_laboratory(text):
    return text.replace("office", "laboratory")

def replace_laboratory_with_research(text):
    return text.replace("laboratory", "research")

def replace_research_with_commercial(text):
    return text.replace("research", "commercial")
def replace_commercial_with_industrial(text):
    return text.replace("commercial", "industrial")
```

```python
appended_df['Question*1.0'] = appended_df['Question*1.0'].apply(remove_numberings)
appended_df.iloc[4:, 0] = appended_df.iloc[4:, 0].apply(replace_warehouse_with_office)
appended_df.iloc[8:, 0] = appended_df.iloc[8:, 0].apply(replace_office_with_laboratory)
appended_df.iloc[12:, 0] = appended_df.iloc[12:, 0].apply(replace_laboratory_with_research)
appended_df.iloc[16:, 0] = appended_df.iloc[16:, 0].apply(replace_research_with_commercial)
appended_df.iloc[20:, 0] = appended_df.iloc[20:, 0].apply(replace_commercial_with_industrial)
appended_df
```

```python


# Iterate over the rows of the first column and modify the entries
appended_df.iloc[:, 0] = [f"{i+1}. {entry}" for i, entry in enumerate(appended_df.iloc[:, 0])]
appended_df
```

```python
appended_df.to_csv('New_weights.csv', index=False)
```
