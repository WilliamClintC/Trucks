```python
import pickle

# Step 1: Define your data
data = [
    {
        "text": "Paragraph 1",
        "section": "Section 1",
        "zoning_title_location": "Location 1"
    },
    {
        "text": "Paragraph 2",
        "section": "Section 2",
        "zoning_title_location": "Location 2"
    },
    {
        "text": "Paragraph 3",
        "section": "Section 3",
        "zoning_title_location": "Location 3"
    }
]

# Step 2: Serialize the data and write it to a file
with open('data.pkl', 'wb') as file:
    pickle.dump(data, file)

# Step 3: Later, when you need to read the data, deserialize it from the file
with open('data.pkl', 'rb') as file:
    loaded_data = pickle.load(file)
    print(loaded_data)
```

```python
import re
import pickle

def extract_and_save_sections(filename):
    # Regular expression pattern to match
    pattern = re.compile(r"\d+\.\d+\.\d+")
    
    # Initialize variables
    sections = []
    current_section = None
    collecting = False
    
    # Open and read the file
    with open(filename, 'r') as file:
        for line in file:
            # Check if the line matches the pattern
            match = pattern.search(line)
            if match:
                # If we were already collecting a section, save it before starting a new one
                if collecting:
                    sections.append(current_section)
                # Start a new section
                current_section = {"section": match.group(), "text": "", "zoning title location": filename}
                collecting = True
            elif collecting:
                # Add the current line to the section's text
                current_section["text"] += line
    
    # Save the last section if there was one being collected
    if collecting:
        sections.append(current_section)
    
    # Save the sections to a pickle file
    with open('sections.pkl', 'wb') as pkl_file:
        pickle.dump(sections, pkl_file)

# Call the function with the filename
extract_and_save_sections('AK_Homer.txt')
```
